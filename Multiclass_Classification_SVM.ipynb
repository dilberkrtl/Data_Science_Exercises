```python
# Importing necessary libraries for dataset creation and visualization
from sklearn.datasets import make_blobs
import mglearn
```

```python
%matplotlib inline
```

```python
# Generating a synthetic dataset of 2D points with 3 classes using make_blobs function
x, y = make_blobs(random_state=42)
```

```python
# Plotting the 2D points with different colors for each class using mglearn's discrete_scatter function
import matplotlib.pyplot as plt
mglearn.discrete_scatter(x[:, 0], x[:, 1], y)

# Setting labels for the x and y axes and adding a legend to identify classes
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.legend(['class 0', 'class 1', 'class 2'])
```

```python
# Importing LinearSVC from sklearn to train a linear SVM model
from sklearn.svm import LinearSVC
```

```python
# Creating a linear SVM classifier and fitting it to the generated data
linear_svm = LinearSVC().fit(x, y)
```

```python
# Extracting the coefficients of the linear SVM (which defines the decision boundaries between classes)
linear_svm.coef_
```

```python
# Visualizing the decision boundaries of the SVM model
import numpy as np

# Re-plotting the data points again for visualization
mglearn.discrete_scatter(x[:, 0], x[:, 1], y)

# Creating a line to plot decision boundaries for each class
line = np.linspace(-15, 15)

# Looping through the coefficients and intercepts for each class to plot the decision boundaries
# For each class, a line is drawn corresponding to the decision boundary
for coef, intercept, color in zip(linear_svm.coef_, linear_svm.intercept_, ['b', 'r', 'g']):
    plt.plot(line, -(line * coef[0] + intercept) / coef[1], c=color)

# Setting the limits of the plot for better visibility of decision boundaries
plt.ylim(-10, 15)
plt.xlim(-10, 8)

# Adding labels and legend to the plot to explain the decision boundaries and the classes
plt.xlabel('Feature 0')
plt.ylabel('Feature 1')
plt.legend(['class 0', 'class 1', 'class 2', 'Decision boundary for class 0', 'Decision boundary for class 1', 'Decision boundary for class 2'], loc=(1.02, 0.4))
```
